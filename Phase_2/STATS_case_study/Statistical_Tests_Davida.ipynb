## 8. Statistical Inference


- Create a 95% confidence interval for the population mean of physically ill days.

```python
df['PHYSHLTH'].mean()
```

```python
import scipy.stats as stats
import numpy as np
import scipy as sp
```

```python
df['PHYSHLTH'].mean()
```

```python
x_bar = np.mean(df['PHYSHLTH'])
s = np.std(df['PHYSHLTH'], ddof = 1)
print(x_bar, s)
```

```python
stats.t.interval(alpha = 0.95,                              # Confidence level
                 df= len(df['PHYSHLTH'])-1,             # Degrees of freedom
                 loc = x_bar,                               # Sample mean
                 scale = s)                             # Standard deviation estimate
```

```python
#Realistically can't have negative sick days we are 95 confident that the population has a sick day number 0 and 23.81 
```
