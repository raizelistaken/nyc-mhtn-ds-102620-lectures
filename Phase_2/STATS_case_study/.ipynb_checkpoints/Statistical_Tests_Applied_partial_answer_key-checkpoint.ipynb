## 8. Statistical Inference


- Create a 95% confidence interval for the population mean of physically ill days.

```python
print(df['PHYSHLTH'].mean())
print(len(df['PHYSHLTH']))
```

```python
df['PHYSHLTH'] #sick days
```

```python
#DO THIS
import scipy.stats as stats
import math
x_bar = np.mean(df['PHYSHLTH']) # mean of vector
s = np.std(df['PHYSHLTH']) # std of vector
n = len(df['PHYSHLTH']) # number of obs
z = 1.96 # for a 95% CI

lower = x_bar - (z * (s/math.sqrt(n)))
upper = x_bar + (z * (s/math.sqrt(n)))
med = np.median(df['PHYSHLTH'])

print(lower, med, upper)

#95percent condifedent the population mean is between 4.98 and 5.3 
```

95 % confident the population mean is between 4.98 and 5.3 

-  Create a 95% confidence interval for the true proportion of chronically sick people.

```python
#DO THIS

samp_prop = df['CHRONIC'].mean()

MoE = 1.96*math.sqrt((samp_prop*(1-samp_prop))/len(df['CHRONIC']))

lower = samp_prop - MoE
upper = samp_prop + MoE

lower, upper
```

# Is there a statistically significant difference in the percentage of men and women who are chronically sick?

Null Hypothesis: p(Men) = p(Women) in terms of chronically ill proportions, Alternative Hypothesis: p(Men) does not = p(Women) in terms of chronically ill proportions
Z-test
1.96 critical val

```python
#4. Evaluate the test statistic against the critical value:
sample1 = df['CHRONIC'].loc[df['SEX'] == 'F']
sample2 = df['CHRONIC'].loc[df['SEX'] == 'M']

x1 = sum(sample1)
x2 = sum(sample2)
n1 = len(sample1)
n2 = len(sample2)
p_star = (x1+x2)/(n1+n2)
p1 = sample1.mean()
p2 = sample2.mean()
z_star = (p1-p2)/math.sqrt(p_star*(1-p_star)*((1/n1)+(1/n2)))
z_star
```

Our test statistic (5.86) is higher than our critical value (1.96). We therefore can reject the null hypothesis. We are 95% confident that we can reject the null hypothesis that the average number of chronically ill men and chronically ill women are equal.

# Perform a statistical test to determine if there is a statistically significant difference in physical health of people from the three states.

```python
CT_nonchronic = df['CHRONIC'].loc[df['_STATE'] == 'Conn'].loc[df['CHRONIC']==0]
CT_chronic = df['CHRONIC'].loc[df['_STATE'] == 'Conn'].loc[df['CHRONIC']==1]
NJ_nonchronic = df['CHRONIC'].loc[df['_STATE'] == 'NJ'].loc[df['CHRONIC']==0]
NJ_chronic = df['CHRONIC'].loc[df['_STATE'] == 'NJ'].loc[df['CHRONIC']==1]
NY_nonchronic = df['CHRONIC'].loc[df['_STATE'] == 'NY'].loc[df['CHRONIC']==0]
NY_chronic = df['CHRONIC'].loc[df['_STATE'] == 'NY'].loc[df['CHRONIC']==1]
```
